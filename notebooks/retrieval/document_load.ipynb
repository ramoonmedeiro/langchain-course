# Loaders com o LangChain
Nesta seção é apresentado formas de você interagir com documentos de diversos formatos e como você pode manipular esses documentos para extrair informações úteis.

## CSV Loader

Como extrair o conteúdo de um arquivo CSV.

```python
from langchain.document_loaders import CSVLoader

csv_file_path = "../../dataset/movies.csv"

csv_loader = CSVLoader(file_path=csv_file_path,  csv_args={'delimiter': ','})
data = csv_loader.load()
```

```python
print(data[0].page_content)
```

## JSON Loader

Como extrair o conteúdo de um arquivo JSON.

```python
from langchain.document_loaders import JSONLoader

json_file_path = "../../dataset/iris.json"

json_loader = JSONLoader(
    file_path=json_file_path,
    #jq_schema="."
    jq_schema=".",
    text_content=False
)

data = json_loader.load()
```

```python
print(data[0].page_content)
```

## PDF Loader

Como extrair o conteúdo de um arquivo PDF.

```python
from langchain.document_loaders import PyPDFLoader

pdf_path = "../../dataset/pdf-TEST.pdf"

loader = PyPDFLoader(file_path=pdf_path)
data = loader.load()
```

```python
print(data[0].page_content)
```

Extraindo imagens de um arquivo PDF.

```python
loader = PyPDFLoader(file_path="https://arxiv.org/pdf/2103.15348.pdf", extract_images=True)
data_image = loader.load()
```
