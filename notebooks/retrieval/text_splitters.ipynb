# Text Splitters

Nesta seção veremos como dividir um texto em partes menores, como palavras, frases, ou parágrafos.

```python
# carregando texto do arquivo pdf
from langchain.document_loaders import PyPDFLoader

pdf_file_path = "https://arxiv.org/pdf/1810.04805.pdf"
loader = PyPDFLoader(file_path=pdf_file_path)
pages = loader.load()
print(pages)
```

```python
len(pages)
```

## Separação por caracteres

```python
from langchain.text_splitter import CharacterTextSplitter

spliter = CharacterTextSplitter(
    separator="\n\n",
    chunk_size=50,
    chunk_overlap=0,
    length_function=len,
    is_separator_regex=False,
)

documents = spliter.split_documents(
    pages
)

documents
```
